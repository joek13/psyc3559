**Defn**: The *validity* of a test or procedure describes the extent to which it measures the phenomenon it's supposed to measure.

#### Four types of validity

**Statistical conclusion validity:** About the correlation (covariance) between the predictor and the outcome
- Does the relationship exist?
- Strictly correlational: non-causal

**Internal validity:** About whether the observed covariation between A (presumed predictor) and B (presumed outcome) actually reflects a causal relationship $A \rightarrow B$

**Construct validity:** About the higher-order constructs that represent sampling particulars
- Given that there is a valid causal relationship, is our interpretation about the construct responsible for that causation reasonable?

**External validity:** Whether the cause-effect relationship holds over variation in persons, settings, treatment variables, and measurement variables.
- Roughly, generalizability: do our results generalize outside of this experimental setting?

#### Threats to different types of validity
Threats to **statistical conclusion validity**: reasons why inferences about covariation between two variables may be incorrect

e.g., number of kids & anxiety during COVID

1. Low statistical power
    - Low power might arise from small sample size, for example
    - Lower sensitivity, in a sense; we won't be able to detect true correlations because of an insufficient sample size
    - More on this on days 3 and 4
2. Violated assumptions of statistical tests
    - It might just be inappropriate to use the tests we try
3. Unreliability of measures
    - Measures of anxiety, for example, might have a low level of reliability
        - Asking participants to self-report these things is notoriously unreliable
4. Restriction of range
    - What if we pose the question as "0 kids", "1-3 kids", "3+ kids"
    - Sometimes misgrouping when you turn a numerical variable into a categorical one is dangerous
5. Unreliability of treatment implementation
    - The experimenters don't apply treatments reliably
6. Heterogeneity of units
    - Different conditions for each experimental unit (subject) that confound our results
    
Threats to **internal validity:** reasons why the causal assessment may be incorrect

E.g., sleep hours & anxiety in COVID

1. Ambiguous temporal precedence
    - Causes always precede effects
    - A lack of clarity about which variable "occurs" first
2. Selection
    - Some factor impacts what kind of respondents we get
    - Confounded by some external factor
3. History
    - Confounded by some event happening in subjects' history
    - E.g., subject loses a job during sleep study, causes stress
4. Maturation
    - Naturally occurring changes over time
    - Maybe people just get better at sleeping with anxiety over time
5. Attrition
    - Missing data :(
    - How do we deal with missing values in our dataset?
    
    
Threats to **construct validity**: reasons why the interpretation the constructs involved may be incorrect

1. Experimenter-bias effect
    - Need to consider the experimenter's own preconceived notions when the study isn't conducted in a blind fashion
2. Unintended by-products of the treatment

Threats to **external validity:** reasons why the findings might not be generalizable
1. Heterogeneity and representativeness of the sample
    - Maybe your sample just doesn't stand in for the greater population
    
#### Reading activity
Flying phobia treatment study

Some issues with the study:
- People with severe flight anxiety self-selected out (by quitting after panicking), so the study might not generalize to all people with flight phobias.
- Participants might feel pressured to report favorable results after all the work they were put through.
- The second time around on the posttest, participants knew for sure that the plane would not actually take off in the simulation.
- Small sample
- People who think their anxiety is treatable self-select in 
- Not clear if pre-flight anxiety is actually a good indicator
- Anxiety criterion is standardized on college students, and participants in the study seem to be a good deal older
- Participants who felt the therapy wasn't working self-selected out by leaving the study 
- The first experience might be artificially high-anxiety because of the panicking subjects (who were absent for the posttest)
- Study claims to be comparative, but ends by making claims about blanket effectiveness

In general: *missing data* can introduce bias if there are different rates of missing data across groups.
Abstracts that reveal researcher expectations can be dangerous.
