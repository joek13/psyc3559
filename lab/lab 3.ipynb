#### Sample size problem
Statistical significance vs. how big the effect of the predictor is

Suppose we randomly take a large national sample of COVID patients and discover their IQ is 101. We also randomly collect a sample of non-COVID Americans whose mean IQ is 101.33. Is this a significant difference? Does it represent a breakthrough finding?

Specifically, suppose the following:

For COVID:
$$ n_1 = 50000 \\ \bar{Y}_1 = 101 \\ s^2_1 = 225 $$
For non-COVID:
$$ n_2 = 50000 \\ \bar{Y}_2 = 101.33 \\ s^2_2 = 225 $$

We conduct a t-test, and we conclude that there is a statistically significant difference. ($p << 0.05$)

"Sample size problem:" a large sample size allows us to conclude a statistically significant result that is practically insignificant

#### R example
Cynthia shows us linear model, with some minor random noise. But the `lm` function only detects statistically significant relationships for sample sizes of $N \approx 50$.

The $N$ we need to pick up on some relationship varies with the scale of the relationship.

This introduces some ambiguity; sometimes, it's unclear whether a statistically significant relationship is the result of a practically significant phenomenon or just a large sample size. For this reason, the consensus has shifted toward stating the results of a hypothesis test alongside an indication of the magnitude of an effect.

The effect size can be stated as a direct difference, or it can be normalized with respect to standard deviation. For example: comparing the means of two groups, an unstandardized effect size would be $\mu_2 - \mu_1$, and a standardized effect size is $\frac{\mu_2 - \mu_1}{\sigma}$, where $\sigma$ is the common standard deviation between the two groups. 

#### Effect size for linear regression
For linear regression, Cohen suggets an effect size $f^2$:
$$ f^2 = \frac{R^2}{1-R^2} $$

When we want to compare extended vs. reduced models, we compute the effect size as follows:

$$ f^2 = \frac{R^2_{full}-R^2_{reduced}}{1-R^2_{full}} $$

#### Statistical power analysis
- Statistical power analysis helps us decide how large of a sample is needed to detect a certain effect reliability

|              | $H_0$ true    | $H_0$ false   |
|--------------|---------------|---------------|
| Don't reject $H_0$ | True positive | Type II error |
| Reject $H_0$ | Type I error  | True negative |

$$P(\text{Type I error}) = P(\text{false positive}) = P(p < \alpha | H_0 \text{ true})$$
$$P(\text {Type II error}) = P(\text{false negative}) = P(p > \alpha | H_0 \text{ false})$$

#### Factors influencing power
- Sample size
- Significance level
- True value of the thing we're testing

#### Why is it missing?
- MAR - Missing at random
    - Missingness is related to the observed data, but not due to the missing data or the latent variables.
- MCAR - Missing completely at random
    - Data are missing for reasons unrelated to the observed data, missing data, or latent variables.
    - Observed data can be treated as a random sample of the complete data.
- MNAR - Missing not at random
    - Missing data is due to unmeasured/unobsered variables
    - We can't account for which data is missing
    
#### Choosing which mechanism
MCAR vs MAR **can be distinguished**: we fit a logistic regression model with DV binary Missing or NotMissing at time step $t$ and IV all the variables collected up until $t$.

For our project, we can treat missing data as MCAR.
