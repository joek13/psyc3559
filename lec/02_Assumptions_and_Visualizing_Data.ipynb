#### p-values
What is the p-value equal to?

Where $X$ is a random variable distributed according to some distribution, and $x$ is our test statistic:

One-sided hypothesis test: $p = P(X \le -|x|)$

Two-sided hypothesis test: $p = 2P(X \le -|x|)$

$\alpha$, the significance level is the cutoff we use to determine significance.

If $p < \alpha$, we conclude significance.

The critical value of the test statistic corresponds with a p-value of $\alpha$. Test statistic values more extreme than the critical value are considered significant, because they correspond with p-values less than $\alpha$.

For this course, $\alpha = 0.05$

#### a p-value is not...
- The probability that the null is true
- The probability that the alternative is false
- The probability that the test statistic is more extreme than the cutoff value
- The probability that the null is true, given that the test statistic is more extreme than the cutoff value
- The probability of obtaining your data, given that the null is true

### Data visualization
#### Why do we need to visualize data?
- Check visually whether data violates assumptions
- So we can more easily find and understand relationships across variables

#### What is an assumption?
- In terms of hypothesis testing, it's a condition required for carrying out a particular hypothesis-testing procedure

#### How do we check assumptions with visual plots?
Checking for normality
- Histograms, QQ plots, box plots

Homogeneity of variance
- Directly: residual plots (when performing models)
- Indirectly: histograms, boxplots (looking at groups)

### Plots
#### Histogram
- Visually check the distribution of numeric data
- Use the `hist` function
- We can visually diagnose...
    - Skew by checking symmetry (long left tail = left/negative skew, long right tail = right/positive skew)
    - Bimodality by looking at peaks
    

#### QQ Plot
- Compare expected values with actual values
- The closer the points are to the dotted diagonal line, the less likely it is that the assumption of normality has been violated
- Visually diagnose...
    - Skew - downward curve = negative skew, upward curve = positive skew
    - Bimodality - the curve goes in two different directions

#### Boxplot
- Boxplots allow you to look at both the distribution and approximate variation of each group
- Visually diagnose
    - Skew - one of the "whiskers" is significantly longer than the other
- Can't easily visually diagnose bimodality... it gets masked

#### Correlation vs regression

#### Regression with continuous dependent variables
- We want to use a variable to predict another variable.
- Sometimes, we want to use more than one predictors to predict another variable
- Our goal is to understand what kinds of changes in one or more variables explain change in the predicted variable

#### Correlation
- Describes the relationship between two numeric ordered variables, that should have *interval* or *ratio* level of measurement
    - If you're using a Pearson correlation coefficient
- Sometimes, the variables are ordinal level
    - This may or may not be problematic statistically, you need to use a *polychoric correlation*
- The correlation coefficient is denoted by the letter $r$

##### Correlation $\ne$ Causation
Spurious correlations are often caused by a third, unmeasured variable

#### Info on $r$

$-1 \le r \le 1$

$r < 0 \rightarrow$ negative relationship

$r >0 \rightarrow$ positive relationship

$$r = \frac{\sum{Z_x * Z_y}}{N - 1}$$

Pearson correlation only quantifies the direction and strength of linear relationships

Curvilinear relationships can be strong and still result in small values of $r$

Testing for correlational significance

Assumptions:
- Population follows normal distribution
- Constant variance at all values of the predictor
    - AKA "homoscedasticity"
    
#### Strong correlation still doesn't imply causation
- Strong correlation could mean...
    - X causes Y
    - Y causes X
    - Some variable Z causes X and Y
    
- Even without causal evidence, we can use regression to make predictions based on predictor data

We're mainly looking at using lines as curves of best fit for bivariate data, and planes as curves of best best fit for multivariate data

$\hat{Y} = b_0 + bX$

$\hat{Y}$, the predicted value of Y

$b_0$, the intercept

$b$, regression coefficient (or slope)

$X$, the predictor variable

$\hat{Y} = b_0 + b_1 X_1 + b_2 X_2 + \dots + b_k X_k$

$b_i$, the partial regression coefficient (weight) for the $i$th predictor variable

#### Quantifying strength of prediction
- We can measure how well our regression line fits to our data by measuring residuals
- Residual is difference between $Y$ and $\hat{Y}$
- If our residuals are small, that's good
- If we find small residuals, our predictor is a good one

Want to minimize the sum of squared errors (OLS regression)

$SS_{error} = \sum{(Y - \hat{Y})^2} = \sum{(Y - (b_0 + bX))^2}$

If we center our line around the means of X and Y, $SS_{error}$ will be minimzed

Bivariate setting:
$$b = \frac{SS_{xy}}{SS_x}=\frac{\sum{(X-\bar{X})(Y-\bar{Y})}}{\sum{(X-\bar{X})^2}}$$

## Correlation and regression
