### Presentation: Three Applications of Regression Analysis: Description, Prediction, Explanation/Model Building

#### Multiple Regression
1. Predictor variables — continuous or categorical
    a. individual differences — personality traits, sex, religion
    b. experimental treatments (e.g., placebo, 1 dose of vax, 2 doses of vax)
    
2. Dependent (criterion) variable
    a. Typically a continuous variable (college GPA, reaction time)
    b. Other forms of DV — extensions of regression
        - binary: e.g., alive or dead
        - ordered categories
        - counts

3. Relationship between independent (X) and dependent (Y) variables

Simple linear relationship: $\hat{Y} = b_0 + b_1X$

#### Three topics
- Description — summarize relationships graphically or in terms of equation
- Prediction — predict a future value for a person based on prior regression equation and this person's predictor values
- Explanation — prior theory and research are used to build a regression model, then tested against the data

These three applications are distinct, but can often be related.

#### Model building
The goal: build a model that predicts odds of mask-wearing.

Some predictors:
- **Conscientiousness:** personality measure; the tendency to be careful, self-disciplined, rule-following
- **Perceived Susceptibility:** health beliefs & health protective behavior; belief about whether we are susceptible
- **Perceived Benefits:** health beliefs; belief about whether an intervention is beneficial

Some criteria and their appropriate models:
- Continuous criterion: OLS regression
- Binary criterion: logistic regression, probit regression
- Count criterion: poission regression, negative binomial regression

How do we evaluate models?
Look at your $R^2$ for the mupltiple regression model. $R^2$, the coefficient of determination, corresponds with the proportion of variation in our criterion explained by our predictors.

Our $R^2_{multiple} = R^2_{Y,\hat{Y}}$ — the correlation between our predictor and observed variable

Model 1: Does conscientiousness explain differences in mask wearing?

Model 2: Do health beliefs about susceptibility explain difference in mask wearing, above and beyond conscientiousness?

Interestingly, very dangerous diseases for which there are no cure show a curvilinear relationship with respect to health beliefs: people would rather just not know they have breast cancer.

Next step: try to replicate model on other datasets; see if predictions are borne out

In our example, we find that our hypothesis about curvilinear relationships w/r/t susceptibility aren't borne out. We take the squared term out.
