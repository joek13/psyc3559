### Regression on Continuous Dependent Variables

Today's example: predicting the mood of healthcare workers using SHHW data.

Our single-variable, toy example:
$X$ is sleep in January, $Y$ is mood

After running single-variable linear regression, we get:

$\widehat{Mood} = 4.23 + (-0.13)X$

- Both effects are significant!
- ...but the $R^2 \approx 0.01$ — very small.

Let's try a different predictor. Now, $X$ is how much sleep we got last week.

$\widehat{Mood} = 2.15 + (0.16)X$

Significant result, with $R^2 \approx 0.04$ — small to medium.

About 4% of variance in mood can be explained by the amount of sleep reported last week.

#### Multiple regression equation
$b_i$ — the partial regression coefficient for the $i$th predictor variable.
$b_0$ - intercept term

Using both sleep last week and sleep in January, we find that both variables are significant.

$R^2 \approx 0.08$

Model: $\widehat{Mood} = 3.46 - 0.23X_{SleepJan} + 0.20X_{SleepLastWeek}$

Interpretation:
- For a given number of hours of sleep last week, each additional hour of sleep in January decreases mood scores by 0.23
- For a given number of hours of sleep in January, each additional hour of sleep last week increases mood scores by 0.20.

We hold constant the other predictors when we interpret the partial coefficients.

ANOVA $F$-test tells us whether additional variables actually introduce significant predictive power

#### Nested models AIC and BIC
AIC - "Akaike's Information Criterion"

BIC - "Bayesian Information Criterion"
$-2logL + kp$, where $L$ is the likelihood function, $p$ is the number of parameters, and $k = log(n)$ for BIC. A maximum likelihood function is fitted to the data.

Both are useful for comparing the relative usefulness of different models

$BIC > AIC$ for any model because it penalizes models with more parameters
